```python
import serial 
import time
```

```python
ser = serial.Serial('COM5', 115200, timeout = 1)
time.sleep(2)
```

```python
while True:
    cmd = input("Enter Command On/Off").strip().lower()
    if cmd in ['on','off']:
        ser.write((cmd+'\n').encode())
        time.sleep(0.1)
        if ser.in_waiting:
            print(ser.readline().decode(errors='ignore').strip())
    else:
        print("Invalid Command, type 'on' or 'off")
```
